```python


# Assuming 'MFR' is the column containing manufacturer codes
mfr_counts = df['mfr'].value_counts()

# Plotting bar chart for manufacturer counts
plt.figure(figsize=(8, 6))
mfr_counts.plot(kind='bar', color='green')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.title('Distribution of Cereal Manufacturers')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


# Assuming 'Rating' is the column containing ratings
ratings = df['rating']

# Plotting histogram for ratings
plt.figure(figsize=(8, 6))
plt.hist(ratings, bins=10, edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings in Cereal Data')
plt.grid(True)
plt.show()
```

```python


# Assuming 'Name' is the column containing cereal names and 'Rating' is the column containing ratings
plt.figure(figsize=(10, 6))
df.boxplot(column='rating', by='name', vert=False, figsize=(12,8))
plt.xlabel('Rating')
plt.ylabel('Cereal Name')
plt.title('Distribution of Ratings by Cereal Name')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


# Load the CSV file
df = pd.read_csv(r'C:\Users\rjhon\Downloads\cereal (1).csv')

# Assuming 'Name' is the column containing cereal names and 'Sugars' is the column containing sugar content
top_10_sugars = df.nlargest(10, 'sugars')

# Plotting bar chart for top 10 products with highest sugars
plt.figure(figsize=(10, 6))
plt.bar(top_10_sugars['name'], top_10_sugars['sugars'], color='orange')
plt.xlabel('Cereal Name')
plt.ylabel('Sugar Content')
plt.title('Top 10 Cereals with Highest Sugar Content')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
